# GRADIENT DESCENT METHODS FOR LINEAR REGRESSION

## [1] Linear regression
Let 

$$ y = \theta_{0} + \theta_{1}x_{1} + \cdots + \theta_{n}w_{n}$$

where $x_{1},\ldots,x_{n} \in \mathbb{R}$ are the independent variables (regressors), $y \in \mathbb{R}$ is dependent variable and $\theta_{0},\ldots,\theta_{n} \in \mathbb{R}$ are unknown paramters that need to be determineid. Define

\begin{align*}
{\boldsymbol \theta} & = ( \theta_{0} , \theta_{1}, \cdots , \theta_{n} )^{\rm T} \in \mathbb{R}^{n+1}\\
{\bf x} & = (1, x_{1} , \cdots , x_{n})^{\rm T} \in \mathbb{R}^{n+1}, \\
{\bf y} & = (y_{1} , \cdots , y_{m})^{\rm T} \in \mathbb{R}^{m}.
\end{align*}

Then, $ y = {\bf x}^{\rm T}{\boldsymbol \theta}$ and note that ${\boldsymbol \theta} \mapsto y(\boldsymbol \theta)$ is linear function of $\boldsymbol \theta$. Given a dataset $ {\bf z}_{k} = ( {\bf x}_{k},y_{k})$, $k \in \{1,\ldots,m \},$ define

\begin{align*}
Z = (X, {\bf y}), \,\, X  = \begin{bmatrix} {\bf x}_{1}^{\rm T} \\ \vdots \\ {\bf x}_{m}^{\rm T} \end{bmatrix} = \begin{bmatrix}
   1       &  x_{1,1} & \cdots &  x_{1,n} \\
   \vdots  & \vdots   & \ddots &  \vdots  \\
   1       & x_{m,1}  & \cdots &  x_{m,n}  
\end{bmatrix} \in \mathbb{R}^{m \times n+1} \,\,\,\, \mbox{and} \,\,\,\
{\bf y}  = \begin{bmatrix}
    y_{1}  \\
    \vdots \\
    y_{m} \end{bmatrix} \in \mathbb{R}^{m},
\end{align*}
For each $k \in \{1,\ldots, m\}$ let the individual loss function be defined by

$$L_{k}( {\boldsymbol \theta};{\bf z}_{k}) := \frac{1}{2} ( {\bf x}_{k}^{\rm T}{\boldsymbol \theta} - y_{k} )^{2}.$$

The term $r_{k}:= {\bf x}_{k}^{\rm T}{\boldsymbol \theta} - y_{k} $ is often referred as "residual". Note that given ${\bf z}_{k},$ $r_{k}$ is an affine function of ${\boldsymbol \theta}.$ The cost function, i.e. the mean of sum of all the individual losses, is given by

\begin{align*}
L( {\boldsymbol \theta};Z) = \frac{1}{m} \sum_{k=1}^{m} L_{k}( {\boldsymbol \theta};{\bf z}_{k} ) = \frac{1}{2m} \sum_{k=1}^{m} (r_{k})^2 = \frac{1}{2m} \| X {\boldsymbol \theta} - {\bf y} \|_{2}^{2}.
\end{align*}

We seek to solve the following least-square problem (LSP):
\begin{align*}
{\boldsymbol \theta}^{*} = \arg \min_{ {\boldsymbol \theta} \in \mathbb{R}^{n+1}}  L( {\boldsymbol \theta};Z).
\end{align*}

This problem is often referred as ``linear least squares`` due to the fact that the residuals are an affine function of the uknown ${\bf w}$. In the data fitting literature, this problem is referred to as a ``regression problem``. Any vector ${\bf w}^*$ that satisfies $ \| X{\boldsymbol \theta}^{*} −{\bf y}\| \leq \| X{\boldsymbol \theta} −{\bf y}\|$ for all ${\boldsymbol \theta}$ is a solution of the least squares problem and we call ${\boldsymbol \theta}^*$ a least squares approximate solution of $X{\boldsymbol \theta} = {\bf y}.$ Clearly, a least squares approximate solution ${\bf w}^*$ of $X{\bf w} = {\bf y}$ need not satisfy the equations $X {\boldsymbol \theta}^{*}= {\bf y}$. However, it makes the the norm of the residual as small as it can be.

The LSP is an unconstrained optimization problem and applying the first order sufficient condition yields the ``normal equation`` given by

\begin{align*}
\nabla_{\boldsymbol \theta} L({\boldsymbol \theta}; Z) = {\bf 0}.
\end{align*}

Before we compute the gradient vector, first note that

\begin{align*}
L( {\boldsymbol \theta};Z) & = \frac{1}{2m} \| X {\boldsymbol \theta} - {\bf y} \|_{2}^{2}\\
              & = \frac{1}{2m}  (X {\boldsymbol \theta} - {\bf y} )^{\rm T} (X {\boldsymbol \theta} - {\bf y} ) \\
                & = \frac{1}{2m}  ({\boldsymbol \theta}^{\rm T}X^{\rm T}X {\boldsymbol \theta} - 2 ( X^{\rm T}{\bf y})^{\rm T} {\boldsymbol \theta} + {\bf y}^{\rm T}{\bf y} ) \\
                & = \frac{1}{2m} \left\{  {\boldsymbol \theta}^{\rm T} A {\boldsymbol \theta} + {\bf b}^{\rm T}{\bf w} + c \right\}
\end{align*}

where $A := X^{\rm T}X \in \mathbb{R}^{n+1 \times n+1},$ ${\bf b} := -2 X^{\rm T}{\bf y} \in \mathbb{R}^{n+1 \times 1},$ and $c := {\bf y}^{\rm T}{\bf y} \in \mathbb{R}_{+}$ with $A = A^{\rm T} \succeq 0.$ Clearly, $L(\cdot)$ is a quadratic function of $\bf w.$ Thus,

\begin{align*}
\nabla_{{\boldsymbol \theta}} L({\boldsymbol \theta}; Z) = \frac{1}{2m} \left\{ 2A {\boldsymbol \theta} + {\bf b} \right\} = \frac{1}{2m} \left\{ 2X^{\rm T}X {\boldsymbol \theta} -2 X^{\rm T}{\bf y} \right\} = \frac{1}{m} X^{\rm T} \left( X {\boldsymbol \theta} - {\bf y} \right).
\end{align*}

The above can be equivalently arrived by noting that

\begin{align*}
\nabla_{{\boldsymbol \theta}} L( {\boldsymbol \theta};Z) = \frac{1}{m} \sum_{k=1}^{m} \nabla_{{\boldsymbol \theta}} L_{k}( {\boldsymbol \theta}; {\bf z}_{k})
\end{align*}

where $\nabla_{{\boldsymbol \theta}} L_{k}( {\boldsymbol \theta}; {\bf z}_{k}) = ( {\bf x}_{k}^{\rm T}{\boldsymbol \theta} - y_{k}) {\bf x}_{k}.$ Now going back to the normal equation, we get

\begin{align*}
\nabla_{{\boldsymbol \theta}} L({\boldsymbol \theta}; Z) = \frac{1}{m} X^{\rm T} \left( X {\boldsymbol \theta} - {\bf y} \right) = {\bf 0} \implies {\boldsymbol \theta}_{LS} = (X^{\rm T}X)^{-1}X^{\rm T}{\bf y}.
\end{align*}

The above requires the invertibility of the matrix $A$ and under some condition $A$ is invertible! 

Finally, to see indeed ${\boldsymbol \theta}_{LS}$ is a minimizer, consider

\begin{align*}
\begin{split}
L( {\boldsymbol \theta};Z) & = \frac{1}{2m}  \| X {\boldsymbol \theta} - {\bf y} \|_{2}^{2} \\
              & = \frac{1}{2m}  \| X ({\boldsymbol \theta}- {\boldsymbol \theta}_{LS}) + X {\boldsymbol \theta}_{LS} - {\bf y} \|_{2}^{2} \\
              & = \frac{1}{2m}  \left\{ \| X ({\boldsymbol \theta}- {\boldsymbol \theta}_{LS})\|_{2}^{2} + \| X {\boldsymbol \theta}_{LS} - {\bf y} \|_{2}^{2} + 2 (X ( {\boldsymbol \theta}-{\boldsymbol \theta}_{LS}))^{\rm T}(X {\boldsymbol \theta}_{LS} - {\bf y}) \right\} \\
              & = \frac{1}{2m}  \| X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS})\|_{2}^{2} +  \frac{1}{2m}  \| X {\boldsymbol \theta}_{LS} - {\bf y} \|_{2}^{2} +  \frac{1}{2m}  2 (X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS}))^{\rm T}(X {\boldsymbol \theta}_{LS} - {\bf y}) \\
              & = \frac{1}{2m}  \| X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS})\|_{2}^{2} +  L( {\boldsymbol \theta}_{LS};Z) +  \frac{1}{m} (X ( {\boldsymbol \theta}- {\boldsymbol \theta}_{LS}))^{\rm T}(X {\boldsymbol \theta}_{LS} - {\bf y})
\end{split}
\end{align*}

Now consider the third term, i.e.

\begin{align*}
\begin{split}
( X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS}) )^{\rm T}(X {\boldsymbol \theta}_{LS}- {\bf y}) & = ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS} )^{\rm T}X^{\rm T}(X {\boldsymbol \theta}_{LS}- {\bf y}) \\
& = {\boldsymbol \theta}^{\rm T} X^{\rm T}X {\boldsymbol \theta}_{LS} - {\boldsymbol \theta}_{LS}^{\rm T}X^{\rm T}X {\boldsymbol \theta}_{LS} - {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} + {\boldsymbol \theta}_{LS}^{\rm T}X^{\rm T} {\bf y} \\
& = {\bf w}^{\rm T} (X^{\rm T}X)  (X^{\rm T}X)^{-1}X^{\rm T}{\bf y} - ((X^{\rm T}X)^{-1}X^{\rm T}{\bf y})^{\rm T}(X^{\rm T}X) (X^{\rm T}X)^{-1}X^{\rm T}{\bf y} - {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} + ((X^{\rm T}X)^{-1}X^{\rm T}{\bf y})^{\rm T}X^{\rm T} {\bf y} \\
& = {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} - ((X^{\rm T}X)^{-1}X^{\rm T}{\bf y})^{\rm T}(X^{\rm T}X) (X^{\rm T}X)^{-1}X^{\rm T}{\bf y} - {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} + ((X^{\rm T}X)^{-1}X^{\rm T}{\bf y})^{\rm T}X^{\rm T} {\bf y} \\
& = {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} - {\bf y}^{\rm T}X(X^{\rm T}X)^{-T} X^{\rm T}{\bf y} - {\boldsymbol \theta}^{\rm T} X^{\rm T}{\bf y} + {\bf y}^{\rm T}X(X^{\rm T}X)^{-T}X^{\rm T} {\bf y} \\
& = {\bf 0}
\end{split}
\end{align*}

Hence,
\begin{align*}
\begin{split}
L( {\boldsymbol \theta}; Z ) = \frac{1}{2m}  \| X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS})\|_{2}^{2} +  L({\boldsymbol \theta}_{LS};Z)
& \implies L( {\boldsymbol \theta};Z ) - L( {\boldsymbol \theta}_{LS};Z) = \frac{1}{2m}  \| X ( {\boldsymbol \theta} - {\boldsymbol \theta}_{LS})\|_{2}^{2} \geq 0 \\
& \implies L( {\boldsymbol \theta};Z) \geq L( {\boldsymbol \theta}_{LS};Z).
\end{split}
\end{align*}

Suppose $L( {\boldsymbol \theta};Z ) = L( {\boldsymbol \theta}_{LS};Z)$, i.e.

\begin{align*}
\frac{1}{2m}  \| X ( {\boldsymbol \theta}- {\boldsymbol \theta}_{LS})\|_{2}^{2} = 0 \iff  X {\boldsymbol \theta} = X {\boldsymbol \theta}_{LS} \implies  {\boldsymbol \theta} = {\boldsymbol \theta}_{LS}.
\end{align*}

Thus, the optimal cost is given by
$$ L( {\boldsymbol \theta}_{LS};Z) = \frac{1}{2m} \| X (X^{\rm T}X)^{-1}X^{\rm T}{\bf y}-{\bf y}\|^{2} = \frac{1}{2m} {\bf y}^{\rm T} (I_{m} - X (X^{\rm T}X)^{-1}X^{\rm T}){\bf y}. $$

## Note
The same can be concluded by noting that $A = X^{\rm T}X \succ 0$.

# [2] Gradient descent methods

The gradient descent algorithms are a family of line search algorithms, i.e.

$$ {\boldsymbol \theta}_{k} = {\boldsymbol \theta}_{k} + \eta_{k} {\bf d}_{k} $$

where $\eta_{k} >0 $ and ${\bf d}_{k} \in \mathbb{R}^{n+1}$ are the step size and search direction, respectively, at the $k$th iteration. These are also first-order algorithms as they only deal with gradient vector. The second-order algorithms involve the second derivatives, i.e. Hessian, but they are computationaly expensive although they perform better.

In the context of regression, depending on how much data we use to compute the gradient of the objective function, there are mainly three variants and they are:
-  Batch gradient descent method, 
-  Stoachastic gradient desecnt method,
-  Mini-batch gradient descent method.

# (i) Batch gradient descent (GD) 
Consider a scalar function $f \colon \mathbb{R} \to \mathbb{R}$ and its Taylor series expansion is given by

$$f(x + \delta x )\approx f(x) + f^\prime(x)\delta x.$$

Since we want to decrease the function, letting $\delta x = -\eta \, f^\prime(x),$ $\eta >0$ is a constant, we have

$$ f(x − \eta  \, f^\prime(x))\approx f(x)- \eta  \, f^\prime(x)^2.$$

For small $\eta$, we obtain

$$ f(x−\eta  \, f^\prime(x)) \leq f(x).$$

That is, updating $x$ as $x := x − \eta \, f^\prime(x)$ may reduce the value of $f(x)$ if $f^\prime(x)\neq 0.$ The positive scalar $\eta$ is called the learning rate or step size. Note that a larger learning rate increases the chance of overshooting the global minimum and  oscillating. However, if the learning rate is too small, the convergence can be very slow. In the case of multivariable function $f \colon \mathbb{R}^{n} \to \mathbb{R},$ it is denoted by $\nabla_{\bf x} f({\bf x})$ and the update is given by

$$ {\bf x} := {\bf x} - \eta  \, \nabla_{\bf x} f({\bf x}).$$

In order to compute the best $\bf w$ we need to find the generalized inverse, which is very difficult. As the problem is an optimization problem, we can use the line search method, i.e. gradient descent method. Let

$$ {\boldsymbol \theta}_{k+1} = {\boldsymbol \theta}_{k} - \eta  \, \nabla L( {\boldsymbol \theta}_{k};Z)$$

where $\eta >0$ is a step-size or learning rate and

\begin{align*}
\left.\nabla L( {\boldsymbol \theta};X,{\bf y})\right|_{ {\boldsymbol \theta}_{k}} = \left. \frac{1}{2m} \left\{ 2 X^{\rm T} ( X  {\boldsymbol \theta} - {\bf y}) \right\}\right|_{ {\boldsymbol \theta}_{k}} = \frac{1}{m} X^{\rm T} ( X  {\boldsymbol \theta}_{k} - {\bf y}).
\end{align*}

## Algorithm

-  Initialize $ {\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$
-  Set learning rate $\eta > 0$ and a tolerence $\epsilon > 0$
-  Set $k = 0$
-  While $\| {\boldsymbol \theta}_{k} - {\boldsymbol \theta}_{k-1}\| > \epsilon$ do
    -  $k \leftarrow k+1$
    -  Evaluate ${\bf g}_{k} = \nabla L({\boldsymbol \theta}_{k-1};Z) = \displaystyle{\frac{1}{m} X^{\rm T} ( X  {\boldsymbol \theta}_{k-1} - {\bf y})} \in \mathbb{R}^{n+1}$
    -  ${\boldsymbol \theta}_{k} \leftarrow {\boldsymbol \theta}_{k-1} - \eta \, {\bf g}_{k-1}$
-  end while

## Fact 
If $ {\boldsymbol \theta}_{0}$ is close enough to the optimum and $\eta$ is sufficiently small, then algorithm achieves linear convergence under some sufficient regularity conditions.

Finally,  the GD ${\bf x}_{k+1} = {\bf x}_{k} - \eta \nabla F( {\bf x}_{k})$ can be seen as a discretization of the ODE $\dot{\bf x} = -\nabla F( {\bf x})$. The slow convergence can be attributed to the fact that after the discretization the iterates starts to wander in zig-zag in directions which is not alligned in the minimizer direction. Through the addition of an inertia term to the iteration should help - this is the idea of Polyak's heavy ball method discussed later on.

```python
############### BATCH GRADIENT DESCENT ##########

def gradient_descent(X, y, w0, eta, max_iterations):
    m, n    = X.shape                  # number of samples
 #   w0      = 5*np.random.rand(n,1)    # parameters
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros( (n, max_iterations))
    W[:,:0] = w0
    L[0]    = (1/(2*m))*np.sum( np.square(X.dot(w0) - y.reshape(m,1)) )
    
    for k in range(max_iterations):
        g1   = np.dot( X.transpose(),(X.dot(w0) - y.reshape(m,1))) / m
        w1   = w0 - eta * g1  # update
        c1   = (1/(2*m))*np.sum( np.square(X.dot(w1) - y.reshape(m,1)) )
        L[k], W[:,:k], G[:,:k] = c1, w1, g1
        w0   = w1
        
    return W, L, G

W_GD, L_GD, G_GD = gradient_descent(Xaug, y, w0, eta, max_iterations)
Result.append( ['GD', W_GD, L_GD] )
#w0 = np.arange(-200, 200, 1)
#w1 = np.arange(-200, 200, 1)
#wx, wy = np.meshgrid(w0, w1)

#def func_z(w0, w1, X, y):
#    z = 0
#    for i in range(m):
#        xi = Xaug[i,:]  # row vector
#        yi = y[i]
#        z = z + np.square( xi[0]*wx + xi[1]*wy - yi )
#    return z/2/m

#fig1, ax1 = plt.subplots()
#z = func_z(w0, w1, Xaug, y)
#ax1.contour(wx, wy, z, cmap='jet')
#ax1.set_xlabel(r'w0')
#ax1.set_ylabel(r'w1')
#x_gd, y_gd = W_GD[0,:].reshape(max_iterations-1,1), W_GD[1,:].reshape(max_iterations-1,1)

#ax1.plot(x_gd, y_gd, 'bo')

#for i in range(1, max_iterations-1):
#    ax1.annotate('', xy=(x_gd[i], y_gd[i]), xytext=(x_gd[i-1], y_gd[i-1]),
#                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
#                   va='center', ha='center')

#plt.show()

#fig2 = plt.figure()
#ax2 = Axes3D(fig2)
#surf = ax2.plot_surface(wx, wy, z, edgecolor='none', rstride=1, cstride=1, cmap='jet')
#plt.show()
```

# (ii) Stochastic gradient descent (SGD) 

The drawback of the GD method is that the cost of computation of the gradient vector at each iteration can be expensive depending on the objective function and the size of the data set. To make each step computationally efficient, instead of computing the gradient of exactly (i.e. using all the data) at each iteration, an estimate of the gradient on the basis of a single randomly picked example will be computed. In other words, the difference between GD and SGD is that GD uses all the samples in the training set to calculate the gradient and do a single update for the parameter, while SGD uses only one sample from the training set to calculate the gradient and do the update for the parameter.

Indeed, the random selection of samples makes the process $\{ {\boldsymbol \theta}_{k}\}$ stochastic and in an indeal case one expect it to behave like the grdient vector computed using the batch process despite the randomness introduced through the sampling. Further, it is not requred to keep track of which sample were selected in the previous iterations, which makes this algorithm suitable for ``on-line`` optimization.

## Algorithm
-  Initialize ${\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$  
-  Set learning rate $\eta >0$
-  Set $k = 0$ (number of epochs)
-  while $ \| {\boldsymbol \theta}_k - {\boldsymbol \theta}_{k-1} \| > \epsilon $ do
    -  Reshuffle the data (very important)
    -  ${\boldsymbol \theta} \leftarrow {\boldsymbol \theta}_{k}$
        -  for $i = 1$ to $m$ do
            -  Evaluate ${\bf g} = \nabla L_{i}( {\boldsymbol \theta} ;{\bf z}_{i}) = ( {\bf x}_{i}^{\rm T}{\boldsymbol \theta} - y_{i}){\bf x}_{i}$
            -  ${\boldsymbol \theta} \leftarrow {\boldsymbol \theta} − \eta \, {\bf g}$
        -  end for
    - $k \leftarrow k +1$
    - ${\boldsymbol \theta}_{k} \leftarrow  {\boldsymbol \theta}$
-  end while

## Question
Why reshuffling is important?
If you don’t reshuffle the data fed to SGD, then the gradient at any particular step is the gradient of a fixed function. For instance,
\begin{align*}
{\bf g}_{k} = \nabla_{\bf w} \left(   \frac{1}{m} \sum_{i= (k-1)m+1}^{km} L( {\bf w};{\bf x}_{i},y_{i}) \right)
\end{align*}
which is a biased estimate of $\nabla L(\cdot)$, where $L$ is the cost function corresponding to the complete data (like GD method). To see why this is a biased estimator, consider
\begin{align*} 
\mathbb{E}_{{\bf x}_{i}}[{\bf g}_{k}] & = \mathbb{E}_{{\bf x}_{i}}\left[\nabla_{\bf w} \left(   \frac{1}{m} \sum_{i= (k-1)m+1}^{km} L( {\bf w};{\bf x}_{i},y_{i}) \right) \right] \\
 & = \nabla_{\bf w} \left( \mathbb{E}_{{\bf x}_{i}}  \left[  \frac{1}{m} \sum_{i= (k-1)m+1}^{km} L( {\bf w};{\bf x}_{i},y_{i}) \right] \right)  \\
 & = \nabla_{\bf w} \left(  \frac{1}{m} \sum_{i= (k-1)m+1}^{km} \mathbb{E}_{{\bf x}_{i}}  \left[  L( {\bf w};{\bf x}_{i},y_{i}) \right] \right) \\
 & = \nabla_{\bf w} \left(  \frac{1}{m} \sum_{i= (k-1)m+1}^{km} L( {\bf w};{\bf x}_{i},y_{i}) \right).
\end{align*}
The last step follows because 
$$\mathbb{E}_{x}[g(X)] = \sum_{i} g(x_i) p(X=x_i)$$
which in our case $g(x_i) = L(\cdot)$ and $p(X=x_i) = 1$ as it is deterministic. On the other hand, to see the effect of reshuffling, consider
\begin{align*} 
\mathbb{E}_{ i \sim U([1,N]) }[{\bf g}] & = \mathbb{E}_{i \sim U([1,N])} \left[\nabla_{\bf w} \left(   \frac{1}{m} \sum_{j = 1}^{m} L( {\bf w};{\bf x}_{j},y_{j}) \right) \right] \\
 & = \nabla_{\bf w} \left( \mathbb{E}_{ i \sim U([1,N])}  \left[  \frac{1}{m} \sum_{j= 1}^{m} L( {\bf w};{\bf x}_{j},y_{j}) \right] \right)  \\
 & = \nabla_{\bf w} \left(  \frac{1}{m} \sum_{j= 1}^{m} \mathbb{E}_{i \sim U([1,N])}  \left[  L( {\bf w};{\bf x}_{j},y_{j}) \right] \right) \\
 & = \nabla_{\bf w} \left(  \frac{1}{m} \sum_{j= 1}^{m} \left\{ \sum_{i=1}^{N} L( {\bf w};{\bf x}_{i},y_{i})  p(X = x_i) \right\}\right) \\
  & = \nabla_{\bf w} \left(  \frac{1}{m} m \sum_{i= 1}^{N} L( {\bf w};{\bf x}_{j},y_{j})  \frac{1}{N} \right) \\
  & = \nabla_{\bf w} \left(  \frac{1}{N} \sum_{i= 1}^{N} L( {\bf w};{\bf x}_{i},y_{i}) \right)
\end{align*}

## Reference
[1] L Bottou, "Large-scale machine learning with stochastic gradient descent," Proceedings of COMPSTAT'2010, pp 177-186, 2010.

```python
#################Stochastic Gradient Descent############################

def stochastic_gradient_descent(X, y,  w0, eta=0.01, max_iterations=10):
    m, n = X.shape               # number of samples
#    w0 = 5*np.random.rand(n,1)    # parameters 
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    W[:,:0] = w0

    for k in range(max_iterations):
        cost = 0.0
        #w = w0
        # Reshuffle the data (IMPORTANT!)
        r = np.random.permutation(m)
        X, y = X[r], y[r]
        
        for i in range(m):
            xi = X[i,:].reshape(n,1)     # column vector
            yi = y[i]                    # scalar
            g1 = (sum(xi[:]*w0[:]) - yi)* xi
            w1 = w0 - eta * g1
            cost += (1/2)*(np.square( sum(xi[:]*w1[:]) - yi))
            w0 = w1

        L[k]  = cost.ravel()/m
        G[:,:k] = g1
        W[:,:k] = w1
        
    return W, L, G

W_SGD, L_SGD, G_SGD = stochastic_gradient_descent(Xaug, y,  w0, eta, max_iterations)
Result.append( ['SGD', W_SGD, L_SGD] )
```

# (iii) Mini-batch Gradient Descent (MB-GD)

The problem with SGD is that the gradient vector with a single sample tends to be very noisy. To avoid this problem, in mini batch algorithm instead of using the complete data set, in every iteration we use a set of $b$ training examples called \emph{batch} to compute the gradient of the cost function and update the parameters. In this method, we create batches from our training data and train our model on that smaller batch, updating the parameter vector all the while.

## Algorithm
-  Initialize ${\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$
-  Set number of epochs, $N \in \mathbb{N}$
-  Set learning rate $\eta > 0$
-  Set a batch size $b$
-  Set $i =1$
-  ${\boldsymbol \theta} \leftarrow {\boldsymbol \theta}_{0}$
-  for $k = 1$ to $N$ do
    -  Evaluate ${\bf g} = \displaystyle{ \frac{1}{b} \sum_{k = i}^{i+b -1} \nabla L_{k}( {\boldsymbol \theta};{\bf z}_{k})}$
    -  ${\boldsymbol \theta} \leftarrow {\boldsymbol \theta} − \eta \, {\bf g}$
    -  $i \leftarrow i + b$
-  end for

```python
############### Mini-batch Stochastic Gradient Descent #########################

def minibatch_stochastic_gradient_descent(X, y, w, eta=0.01, max_iterations=10):
    m, n = X.shape
    L, W, G = np.zeros(max_iterations), np.zeros( (n, max_iterations-1)), np.zeros( (n, max_iterations-1))
 #   w = 5*np.random.rand(n,1)
    W[:,:0] = w
    b = 20
    n_batches = int(m/b)
   
    for k in range(max_iterations):
        cost = 0.0
        r = np.random.permutation(m)
        X, y = X[r], y[r]

        for j in range(0, m, b):
            Xi = X[j:j+b]
            yi = y[j:j+b]
            p, q = Xi.shape
            g = np.dot( Xi.transpose(), Xi.dot(w) - yi.reshape(p,1)) / p
            w = w - eta*g
            cost += (1/(2*p))*np.sum( np.square(Xi.dot(w) - yi.reshape(p,1)) )

        L[k] = cost.ravel()/n_batches
        G[:,:k] = g
        W[:,:k] = w
        
    return W, L, G


W_MBGD, L_MBGD, G_MBSGD = minibatch_stochastic_gradient_descent(Xaug, y,  w0, eta, max_iterations)
Result.append( ['MB-SGD', W_MBGD, L_MBGD] )
```

# (a) Averaged SGD (A-SGD)

Unlike the usual SGD, this method keeps track of all the previous gradient vectors, and after the optimization the averaged parameter vector takes the place of $\bf w$, i.e. at the $k$th step

\begin{align*}
\overline{\bf w}_{k} := \frac{1}{k} \sum_{i=1}^{k} {\bf w}_{k} .
\end{align*}

Note that the above can be written recursively as

\begin{align*}
\overline{\bf w}_{k} = \frac{1}{k} \left\{ \sum_{i=1}^{k-1} {\bf w}_{i} + {\bf w}_{k} \right\}  = \left( \frac{k-1}{k} \right) \frac{1}{k-1} \sum_{i=1}^{k-1} {\bf w}_{i} + \frac{1}{k}  {\bf w}_{k}  = \left( \frac{k-1}{k} \right) \overline{\bf w}_{k-1} + \frac{1}{k}  {\bf w}_{k}. 
\end{align*}

## Algorithm
-  Initialize ${\bf w}_{0} \in \mathbb{R}^{n+1}$
-  Set number of epochs, $N \in \mathbb{N}$
-  Set learning rate $\eta >0$
-  for $k = 1$ to $N$ do
    -  Reshuffle the data (very important)
    -  ${\bf w}_{0} \leftarrow \overline{\bf w}_{m}$
        -  for $i = 1$ to $m$ do
            -  Evaluate ${\bf g}_{i} = \nabla L_{i}({\bf w}_{i-1};{\bf z}_{i}) = ( {\bf w}_{i-1}^{\rm T}{\bf x}_{i}-y_{i}){\bf x}_{i}$
            -  ${\bf w}_{i} \leftarrow {\bf w}_{i-1} − \eta \, {\bf g}_{i-1}$
            -  $\displaystyle{\overline{\bf w}_{i} = \left( \frac{i-1}{i} \right) \overline{\bf w}_{i-1} + \frac{1}{i}  {\bf w}_{i}}$
        -  end for
-  end for


## Reference
[1] BT Polyak and AB Juditsky, "Acceleration of stochastic approximation by averaging," SIAM Journal of Control and Optimization, 30(4):838-855, 1992.

```python
###################### Average Stochastic Gradient Descent ###############

def average_stochastic_gradient_descent(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape                    # number of samples
    w_avg = 5*np.random.rand(n,1)     # parameters 
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    W[:,:0] = w_avg
    
    for k in range(max_iterations):
        cost = 0.0
        w = w_avg
        # Reshuffle the data (IMPORTANT!)
        r = np.random.permutation(m)
        X, y = X[r], y[r]
        
        for i in range(1, m ):
            xi = X[i,:].reshape(n,1)     # row vector
            yi = y[i]                    # scalar
            g = (sum(xi[:]*w[:]) - yi)* xi
            w = w - eta * g
            w_avg = (i-1)*w_avg/i + w/i
            cost += (1/2)*(np.square( sum(xi[:]*w[:]) - yi))

        L[k]  = cost.ravel()/m
        G[:,:k] = g
        W[:,:k] = w_avg
        
    return W, L, G

W_ASGD, L_ASGD, G_ASGD = average_stochastic_gradient_descent(Xaug, y, eta, max_iterations)
Result.append( ['A-SGD', W_ASGD, L_ASGD] )
```

# (b) SGD with momentum (M-SGD) method

For a function ${\bf x} \mapsto F( {\bf x}),$ it is well-known that the update term $\Delta {\bf x}_{k} = \eta \nabla F( {\bf x}_{k})$ in GD method makes it slower. Large values of $\eta$ can diverge the GD method, while small value can make it very slow. In particular, if there is a kind of 1D shallow valley, then the gradient method can oscillate whereby making its convergence slower. The idea of momentum method is to reduce the variation in the vertical and make faster in horizontal direction. 

First, we will recall the exponential weighted average. Given a noisy measurement sequence ${\bf y}_{k},$ we would like to have a moving average which would remove the effect of noise and approximate it closer to the original function. To this effect, define a new squence ${\bf v}_{k}$ given by

\begin{align}
{\bf v}_{k} & = \gamma \, {\bf v}_{k-1} + (1 - \gamma) \, {\bf y}_{k},\,\,\, k = 1,2 , \ldots, \\
{\bf v}_{0} & = \mu_{0},
\end{align}

where $\gamma \in [0,1],$ i.e. ${\bf v}_{k}$ is a convex combination of ${\bf v}_{k-1}$ and ${\bf y}_{k}.$ Combining the above, we have

\begin{align} 
{\bf v}_{k} = \gamma \, {\bf v}_{0} +  (1-\gamma) \sum_{j=1}^{k} \gamma^{k-j} {\bf y}_{j}.
\end{align}

In the above equation, notice that at the $k$th instance, ${\bf v}_{k}$ is a linear combiniatin of $y_{j}$'s from 1 to $k$ with the coefficient $(1-\gamma)\gamma^{k-j} \in [0,1].$ Since $0 \leq \gamma \leq 1,$ it follows that for larger value of $k$ it diminishes rapidly thereby the contribution of older values of ${\bf y}_{j}$ is negligible at the $k$ the instance. A good choice of value for this approximation is the weight becomes smaller than $1 / e.$ Larger the $\gamma$ means averaging over over larger number of data points. Further, in the beginning, there are not enough samples to take the average. This results in poor averages. To circumvent this, in $\bf v$ we can use "bias-corrected" verion of $\bf v$, i.e 

$$ {\bf v}^{\prime}_{k} = \frac{ {\bf v}_{k}}{ 1 - \beta^k}.$$

As $k \to \infty$, $1 - \beta^k \to 1$ and  ${\bf v}^{\prime}_{k}  \to {\bf v}_{k}.$ On the other hand, As $k \to 0$, $1 - \beta^k \to 0$ and  ${\bf v}^{\prime}_{k}  \to \infty.$  With bias correction, it produces better results for smaller values of $k$. 

In the GD with momentum method we use the exponetially weighted average of gradients and use it to update the paramter, i.e. 

\begin{align}
{\bf v}_{k} & = \gamma \, {\bf v}_{k-1} + (1 - \gamma) \, \nabla F( {\bf x}_{k}),\,\,\, k = 1,2 , \ldots, \\
{\bf x}_{k+1} & = {\bf x}_{k} - \eta {\bf v}_{k}.
\end{align}


Then the update rule is given by $\Delta {\bf x}_{k} = \eta  \,{\bf v}_{k}$, i.e. $ {\bf x}_{k+1} = {\bf x}_{k} - \eta {\bf v}_{k}.$ 
Expanding recursive relation of the $\bf v$ update we get

\begin{align} 
{\bf v}_{k} &= (1-\gamma)\sum_{i=0}^{k-1} \gamma^{i} \nabla L( {\bf w}_{k-1-i}) = (1-\gamma) \left\{ \gamma^{0} \nabla L( {\bf w}_{k-1}) + \gamma^{1} \nabla L( {\bf w}_{k-2}) + \dots + \gamma^{k-2} \nabla L({\bf w}_1) + \gamma^{k-1} \nabla L({\bf w}_0) \right\} 
\end{align}

Barring the term $(1-\gamma),$ the expression inside the bracket can be viewed as linear combination of all previous gradients with coefficients vector given by $(\gamma^{0}, \gamma^{1}, \dots, \gamma^{k-1}).$ Since $\gamma \in (0,1),$ for larger values of $k$, the corresponding graidents are forgotten exponentially, i.e. there is a diminishing importance towards the older gradients. However, all of the gradients contribute to the current direction of parameter update.

There is anotehr way to write the above by omitting $1-\gamma$ term, i.e

\begin{align}
{\bf v}_{k} & = \gamma \, {\bf v}_{k-1} + \eta \, \nabla F( {\bf x}_{k}),\,\,\, k = 1,2 , \ldots, \\
{\bf x}_{k+1} & = {\bf x}_{k} - {\bf v}_{k},
\end{align}

which is identical to the first set of equations except that one needs to scale $\eta$ by $(1 - \gamma).$


## Algorithm
-  Initialize ${\boldsymbol \theta}_{0},{\bf v}_{0} \in \mathbb{R}^{n+1}$
-  Set number of epochs, $N \in \mathbb{N}$
-  Set learning rate $\eta >0$ and the momentum parameter $\gamma \in [0,1]$
-  for $k = 1$ to $N$ do
     -  Evaluate ${\bf g}_{k}$ on the current mini batch
     -  ${\bf v}_{k} = \gamma \, {\bf v}_{k-1} + (1 - \gamma) {\bf g}_{k}$
     -  ${\boldsymbol \theta}_{k+1} \leftarrow {\boldsymbol \theta}_{k} − \eta \, {\bf v}_{k}$
-  end for

## Reference

[1] DE Rumelhart, GE Hinton, and JR Williams, "Learning representations by back-propagating errors". Nature. 323 (6088): 533–536,  1986.

[2] N Qian, "On the momentum term in gradient descent learning algorithms," Neural Networks: The Official Journal of the International Neural Network Society, 12(1), 145–151, 1999. 

```python
############ Momentum Method #############################################

def momentum_stochastic_gradient_descent(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    L, W, G = np.zeros( max_iterations ), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    w0 = 25*np.random.rand(n,1)
    v0 = np.ones( (n,1) )
    gamma = 0.9
    W[:,:0] = w0
 
    for k in range(max_iterations):
        cost = 0.0
        #w_0 = w_m
        #v_0 = v_m
        # Reshuffle the data (IMPORTANT!)
        r = np.random.permutation( m )
        X, y = X[r], y[r]
        
        for i in range( m ):
            xi = X[i,:].reshape(n,1)               # row vector
            yi = y[i]                  # scalar
            g1 = (sum(xi[:]*w0[:]) - yi)* xi #((xi.dot(w_m)-yi)*xi).reshape( (n,1))
            v1 = gamma * v0 + (1-gamma) * g1
            w1 = w0 - eta* v1
            cost += (1/2)*(np.square( sum(xi[:]*w1[:]) - yi))
            w0, v0 = w1, v1

        L[k]  = cost.ravel()/m
#        w_m = w_m - v_m
        G[:,:k] = g1
        W[:,:k] = w1

    return W, L, G

W_MSGD, L_MSGD, G_MSGD = momentum_stochastic_gradient_descent(Xaug, y, eta, max_iterations)
Result.append( ['M-SGD', W_MSGD, L_MSGD] )
```

## (c) Heavy-ball method or Polyak’s momentum algorithm

The heavy ball method is due to Polyak [1] aimed at accelerating the GD method and inspired from the second order ODE describing the motion of a particle. This is also called two-step method. The idea here is to combine current gradient (gravity) and previous search direction (momentum) linearly. 

Consider a function $F \colon \mathbb{R}^{n} \to \mathbb{R},$ i.e. ${\bf x} \mapsto F( {\bf x}).$ Choose initial point ${\bf x}_{0} \in \mathbb{R}^{n}$ and let ${\bf x}_{-1} = {\bf x}_{0}.$ The update rule is given by

\begin{align}
{\bf x}_{k+1}  = \underbrace{ {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k})}_{\rm GD} + \gamma \, \underbrace{ ( {\bf x}_{k} - {\bf x}_{k-1})}_{\rm MOMENTUM},\, k \geq 1.
\end{align}

This can be viewed as a multi-step iterative method which exploits iterates prior to the most recent one, i.e. 

\begin{align}
{\bf x}_{k+1} & =  {\bf x}_{k} + \eta \, \color{red}{ {\bf p}_{k}} \\
\color{red}{ {\bf p}_{k}} & = \beta \, \color{red}{  {\bf p}_{k-1} } - \nabla F( {\bf x}_{k} ) 
\end{align}

for initial conditions ${\bf x}_{0},$ ${\bf p}_{0}$ (typically zero), and parameters $\eta, \beta.$ Combining the two in one, we get

\begin{align}
\begin{split}
{\bf x}_{k+1} & = {\bf x}_{k} + \eta \, \color{red}{ {\bf p}_{k}} \\
              & = {\bf x}_{k} + \eta \, \left\{  \beta \, \color{red}{  {\bf p}_{k-1} } - \nabla F( {\bf x}_{k} ) \right\}  \\
              & = {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k}) + \eta \, \beta \, ( {\bf x}_{k} - {\bf x}_{k-1}) \\
              & = \underbrace{ {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k})}_{\rm GD} + \gamma \, \underbrace{ ( {\bf x}_{k} - {\bf x}_{k-1})}_{\rm MOMENTUM}
\end{split}
\end{align}

where $\gamma = \eta \beta$ and term ${\bf x}_{k} − {\bf x}_{k-1}$ is referred to as momentum. Thus, momentum method is nothing but gradient method with a modified gradient by adding the extra momentum term to the gradient.

This method can be viewed as the motion of a particle under a potential function $F$ and friction described by the 2nd order ODE

$$ \ddot{\bf x} + a \,\dot{\bf x} + b\,  \nabla F({\bf x}) = {\bf 0}$$

where $a,b >0.$ Discretizing the above, gives

$$ ( {\bf x}_{k+1} - 2 {\bf x}_{k} +  {\bf x}_{ k−1})+ a ( {\bf x}_{k+1} - {\bf x}_{k})+ b \nabla F( {\bf x}_k) =0 $$

which can be further reduced to 

\begin{align}
{\bf x}_{k+1} & =  \underbrace{ {\bf x}_{k} - \left( \frac{b}{1+a} \right) \nabla F( {\bf x}_k)}_{\rm GD} + \left( \frac{1}{1+a} \right) \underbrace{ ( {\bf x}_{k} - {\bf x}_{k-1})}_{\rm MOMENTUM} \\ 
              & =  {\bf x}_{k} - \left( \frac{b}{1+a} \right)  \underbrace{ \left\{ \nabla F( {\bf x}_k) + \frac{1}{b} ( {\bf x}_{k-1} - {\bf x}_{k}) \right\}}_{= \nabla \tilde{F}({\bf x})}.
\end{align}

The last equation in the above equation is akin to the GD method but with a different function. To see this, let ${\bf a}$ be a given constant, and define 

$$ \tilde{F}( {\bf x}) =  F( {\bf x}) - \frac{1}{2b} \| {\bf x} - {\bf a}\|^{2}.$$ 

Then, $ \nabla \tilde{F}( {\bf x}) = \nabla F( {\bf x}) + ({\bf a} - {\bf x} )/b.$ In view of this, by letting ${\bf a} = {\bf x}_{k-1}$, we have $ {\bf x}_{k+1} =  {\bf x}_{k} - b(1 + a)^{-1} \nabla \tilde{F}( {\bf x}_k).$

Recall that every $n$th order linear ODE can be equivalently written as $n$ first order ODEs. Let ${\bf x}_{1} ={\bf x}.$ Then,

\begin{eqnarray}
\begin{array}{lcl}
\dot{\bf x}_{1} & = & {\bf x}_{2} \\
\dot{\bf x}_{2} & = & - a {\bf x}_{1} - b \nabla F( {\bf x}_{1})
\end{array} & \iff &  
\begin{array}{lcl}
\begin{bmatrix} 
\dot{\bf x}_{1}  \\ \dot{\bf x}_{2} 
\end{bmatrix}  =  \begin{bmatrix}
0 & 1 \\ -a & 0 
\end{bmatrix}\begin{bmatrix}
{\bf x}_{1} \\ {\bf x}_{2} \\
\end{bmatrix} + \begin{bmatrix}
{\bf 0} \\ - b \nabla F( {\bf x}_{1})
\end{bmatrix}
\end{array}
\end{eqnarray}

## Algorithm

-  Initialize ${\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$
-  Set learning rate $\eta > 0,$ momentum parameter $\gamma >0,$ and a tolerence $\epsilon > 0$
-  Set $k = 0$
-  While $\| {\boldsymbol \theta}_{k} - {\boldsymbol \theta}_{k-1}\| > \epsilon$ do
    -  $k \leftarrow k+1$
    -  Evaluate ${\bf g}_{k} = \nabla L( {\boldsymbol \theta}_{k-1};Z) = \displaystyle{\frac{1}{m} X^{\rm T} ( X  {\boldsymbol \theta}_{k-1} - {\bf y})} \in \mathbb{R}^{n+1}$
    -  ${\boldsymbol \theta}_{k+1} \leftarrow {\boldsymbol \theta}_{k} −\eta \, {\bf g}_{k} + \gamma  \,( {\boldsymbol \theta}_{k} - {\boldsymbol \theta}_{k-1})$
-  end while

### References
[1] BT Polyak, Some methods of speeding up the convergence of iteration methods. USSR Computational Mathematics and Mathematical Physics 4, pp. 1–17, 1964.

```python
############### Heavy ball method ##########

def heavy_ball(X, y, eta, max_iterations):
    m, n = X.shape       # number of samples
    w0 = 5*np.random.rand(n,1)    # parameters
    w1 = 5*np.random.rand(n,1)    # parameters
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    beta = 0.1
    W[:,:0] = w0
    W[:,:1] = w1
    L[0] = (1/(2*m))*np.sum( np.square(X.dot(w0) - y.reshape(m,1)) )
    L[1] = (1/(2*m))*np.sum( np.square(X.dot(w1) - y.reshape(m,1)) )

    for k in range(2,max_iterations):
        g = np.dot( X.transpose(),(X.dot(w1) - y.reshape(m,1))) / m
        w = w1 - eta * g + beta*(w1 - w0) # update
        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
        W[:,:k] = w
        G[:,:k] = g
        w0, w1 = w1, w
        
    return W, L, G


W_HB, L_HB, G_HB = heavy_ball(Xaug, y, eta, max_iterations)
Result.append( ['HB', W_HB, L_HB] )
```

# (d) Nesterov accelerated gradient

In [2] it was shown that there exists a relatively simple convex function and specific set of parameters for which the heavy-ball method fails to converge. This was because the iterates were stuck in a limit cycle. Building on Polyak's idea of momentum or heavy ball method, Nesterov fixed this issue with the same acceleration as the heavy-ball method, but that can be shown to converge for general convex functions. Nesterov’s algorithm has two sequences working together to produce a corrective movement compared to the Polyak's heavy-ball method. 

Consider ${\bf x} \mapsto F( {\bf x}),$ and as in momentum method let

\begin{align}
{\bf x}_{k+1}  = \underbrace{ {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k} + \beta ({\bf x}_{k} - {\bf x}_{k-1}))}_{\rm GD} + \beta \, \underbrace{ ( {\bf x}_{k} - {\bf x}_{k-1})}_{\rm MOMENTUM}.
\end{align}

Comparing this with that of heavy ball update rule, we see that Polyak’s method evaluates the gradient before adding momentum, while Nesterov’s algorithm evaluates the gradient after applying momentum.

\begin{align}
\begin{bmatrix}
{\bf p}_{k}  \\ {\bf x}_{k+1} \end{bmatrix} = \begin{bmatrix}
\beta & {\bf 0}_{2} \\
\eta  & I_{2} \end{bmatrix}\begin{bmatrix}
{\bf p}_{k-1}  \\ {\bf x}_{k} \end{bmatrix} + \begin{bmatrix}
−\nabla F( {\bf x}_{k} + \beta ({\bf x}_{k} - {\bf x}_{k-1}))\\  {\bf 0} \end{bmatrix}
\end{align}

for initial conditions ${\bf x}_{0},$ ${\bf p}_{0}$ (typically zero), and parameters $\eta, \beta.$ Combining the two in one,
we get

\begin{align}
\begin{split}
{\bf x}_{k+1} & = {\bf x}_{k} + \eta \, {\bf p}_{k} \\
              & = {\bf x}_{k} + \eta \, \left\{  −\nabla F( {\bf x}_{k}+ \beta ({\bf x}_{k} - {\bf x}_{k-1}) ) + \beta \, {\bf p}_{k-1} \right\}  \\
              & = {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k}+ \beta ({\bf x}_{k} - {\bf x}_{k-1})) + \eta \, \beta \, ( {\bf x}_{k} - {\bf x}_{k-1}) \\
              & = \underbrace{ {\bf x}_{k} − \eta \, \nabla F( {\bf x}_{k}+ \beta ({\bf x}_{k} - {\bf x}_{k-1}))}_{\rm GD} + \gamma \, \underbrace{ ( {\bf x}_{k} - {\bf x}_{k-1})}_{\rm momentum}
\end{split}
\end{align}

## Algorithm

-  Initialize ${\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$
-  Set learning rate $\eta > 0$, momentum parameter $\mu$, and a tolerence $\epsilon >0$
-  Set a tolerence $\epsilon > 0$
-  Set $k = 0$
-  ${\bf p}_{0} \leftarrow {\boldsymbol \theta}_{0}$
-  While $\| {\boldsymbol \theta}_{k} - {\boldsymbol \theta}_{k-1}\| > \epsilon$ do
    -  $k \leftarrow k+1$
    -  ${\bf p}_{k+1} = {\boldsymbol \theta}_{k} - \gamma \nabla L( {\boldsymbol \theta}_{k-1};Z)$
    -  ${\boldsymbol \theta}_{k+1} = {\bf p}_{k+1} + \mu ( {\bf p}_{k+1} − {\bf p}_{k})$
-  end while

## Reference

[1] Y Nesterov, "A method for unconstrained convex minimization problem with the rate of convergence o(1/k2)," Doklady ANSSSR (translated as Soviet.Math.Docl.), vol. 269, pp. 543– 547, 1983.

[2]  L. Lessard, B. Recht, and A. Packard.  Analysis and Design of Optimization Algorithms via Integral Quadratic
Constraints. ArXiv e-prints, Aug. 2014.

```python
############ Momentum Method #############################################

def nesterov_accelerated_gradient(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    w0 = 5*np.random.rand(n,1)
    p0 = w0
    gamma = 0.9
    W[:,:0] = w0
    L[0] = (1/(2*m))*np.sum( np.square(X.dot(w0) - y.reshape(m,1)) )

    for k in range(1,max_iterations):
        g0 = np.dot( X.transpose(),(X.dot(w0) - y.reshape(m,1))) / m
        p1 = w0 - eta*g0
        w1 = p1 + gamma*(p1 - p0)
        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w1) - y.reshape(m,1)) )
        W[:,:k] = w1
        G[:,:k] = g0
        w0, p0 = w1, p1

#    for k in range(max_iterations):
#        cost = 0.0
#        w_0 = w_m
#        v_0 = v_m
#        # Reshuffle the data (IMPORTANT!)
#        r = np.random.permutation( m )
#        X, y = X[r], y[r]
        
#        for i in range( m ):
#            xi = X[i,:].reshape(n,1)             # row vector
#            yi = y[i]                  # scalar
#            y1 = x0 - eta * g0
#            x1 = y1 + gamma*(y1 - y0)
#            g = (sum(xi[:]*(w_m[:] + gamma* v_m[:])) - yi)* xi #((xi.dot(w_m - gamma* v_m)-yi)*xi).reshape( (n,1))
#            v_m = gamma * v_m + eta * g
#            cost += (1/2)*(np.square( sum(xi[:]*w_m[:]) - yi))
            
#        L[k]  = cost.ravel()/m
#        w_m = w_m - v_m
#        W[:,:k] = w_m
    return W, L, G

W_NAG, L_NAG, G_NAG = nesterov_accelerated_gradient(Xaug, y, eta, max_iterations)
Result.append( ['NAG', W_NAG, L_NAG] )
```

# (e) Fast Iterative Shrinkage-Thresholding Algorithm  (FISTA)

Beck and Teboulle proposed a similar momentum-based algorithm:

-  Choose initial ${\bf x}_{0}$ and let ${\bf y}_{1} = {\bf x}_{0}$, $t_{1} = 1$
-  for $k \geq 1,$ do
    *  ${\bf x}_{k} \leftarrow {\bf y}_{k}-\eta \nabla F( {\bf x}_{k})$
    *  $t_{k+1} \leftarrow \frac{1}{2}\left( 1 + \sqrt{ 1 + 4 t_{k}^{2}} \right)$
    *  ${\bf y}_{k+1} \leftarrow {\bf x}_{k} + \frac{t_{k}-1}{t_{k+1}}({\bf x}_{k} - {\bf x}_{k-1})$
-  end for


```python
############ FISTA #############################################

def FISTA(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    w0 = 5*np.random.rand(n,1)
    y1 = w0
    t1 = 1
    W[:,:0] = w0
    L[0] = (1/(2*m))*np.sum( np.square(X.dot(w0) - y.reshape(m,1)) )

    for k in range(1,max_iterations):
        g0 = np.dot( X.transpose(),(X.dot(w0) - y.reshape(m,1))) / m
        w1 = y1 - eta*g0
        t2 = 0.5*( 1 + np.sqrt(1+ 4*t1**2))
        y2 = w1 + ((t1-1)/t2)*(w1 - w0)
        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w1) - y.reshape(m,1)) )
        W[:,:k] = w1
        G[:,:k] = g0
        t1, w0, y1 = t2, w1, y2

    return W, L, G

W_FISTA, L_FISTA, G_FISTA = FISTA(Xaug, y, eta, max_iterations)
Result.append( ['FISTA', W_FISTA, L_FISTA] )
```

# (f) Root Mean Square Propogation (RMSProp)

RMSProp is a modification of the momentum method, which is used to speed up the convergence by appropriately scaling the updates in different directions by  dividing the gradient by a running average of its recent magnitude. The algorithm can be written as       

\begin{align}
{\bf s}_{k+1} & = \beta \, {\bf s}_{k} + (1-\beta)\, \left( \nabla L({\bf x}_{k}) \circ \nabla L({\bf x}_{k}) \right), \\
{\bf x}_{k+1} & = {\bf x}_{k} - \eta \frac{ \nabla L( {\bf x}_{k})}{ \sqrt{{\bf s}_{k}} },
\end{align}

where 
-  $A \circ B$ denotes the Hadamard product (elementise multiplication) of two same sized matrices $A$ and $B$, i.e. $[A \circ B]_{ij} = [A]_{ij} [B]_{ij},$ for all $1 \leq i \leq m$ and $1 \leq j \leq n,$ and 
-  square root is the elementwise operation, i.e. $\sqrt{ \bf a} = \left( \sqrt{a_1}, \ldots, \sqrt{ a_n} \right).$

In componentwise, the update rule can be described by

\begin{align*}
{\bf e}_{j}^{\rm T}{\bf x}_{k+1} = {\bf e}_{j}^{\rm T}{\bf x}_{k} - \eta \frac{ {\bf e}_{j}^{\rm T} \nabla L( {\bf x}_{k})}{ \sqrt{ {\bf e}_{j}^{\rm T}{\bf s}_{k}} }, \, j=1,2,\ldots, n.
\end{align*}

To gain insight into it, suppose the $i$th component of the gradient is very small. This implies that the corresponding component of $\bf s$ also small. Consequently, in the update step this will make larger contribution, and vice versa. The name because it involves squaring and square root operations.

## Algorithm

-  Initialize ${\boldsymbol \theta}_{0}, {\bf s}_{0} \in \mathbb{R}^{n+1}$
-  Set learning rate $\eta > 0,$ parameter $\beta >0,$ and a tolerence $\epsilon > 0$
-  Set $k = 0$
-  While $\| {\boldsymbol \theta}_{k} - {\boldsymbol \theta}_{k-1}\| > \epsilon$ do
    -  Evaluate ${\bf g}_{k+1} = \nabla L( {\boldsymbol \theta}_{k};Z) = \displaystyle{\frac{1}{m} X^{\rm T} ( X  {\boldsymbol \theta}_{k} - {\bf y})} \in \mathbb{R}^{n+1}$
    -  $ {\bf s}_{k+1} = \beta \, {\bf s}_{k} + (1-\beta)\, \left( {\bf g}_{k} \circ {\bf g}_{k} \right)$
    -  ${\boldsymbol \theta}_{k+1} \leftarrow {\boldsymbol \theta}_{k} −\eta \displaystyle{\frac{  {\bf g}_{k}}{ \sqrt{{\bf s}_{k}} }}$
    -  $k \leftarrow k+1$
-  end while
 
### References
[1] T Tieleman and G Hinton, Lecture 6.5 - RMSProp, COURSERA: Neural Networks for Machine Learning, 2012.

```python
###################### RMSProp ##################################

def RMSProp(X, y, eta=0.01, max_iterations=10):
    eps, beta = 1e-08, 0.999
    m, n = X.shape
    L, W = np.zeros(max_iterations), np.zeros((n,max_iterations-1))
    w0, s0 = 5*np.random.rand(n,1), np.zeros((n,1))
    W[:,:0] = w0
    
    for k in range(max_iterations):
        cost = 0.0
        
        # Reshuffle the data (IMPORTANT!)
        r = np.random.permutation(m)
        X, y = X[r], y[r]
        
        for i in range(m):
            xi = X[i,:].reshape(n,1)           # column vector
            yi = y[i]                          # scalar
            g1 = (sum(xi[:]*w0[:]) - yi)* xi
            s1 = beta*s0 + (1 - beta)*(g1*g1)
            w1 = w0  - eta * g1/np.sqrt(s1)
            cost += (1/2)*(np.square( sum(xi[:]*w1[:]) - yi))
            w0, s0 = w1, s1
            
        L[k] = cost.ravel()/m
        W[:,:k] = w1
        
    return W, L

W_RMSProp, L_RMSProp = RMSProp(Xaug, y, eta, max_iterations)
Result.append(['RMSProp', W_RMSProp, L_RMSProp])
```

# (g) Adaptive gradient (AdaGrad)

Let ${\bf g}_{k}$ denote the partial derivative of the objective function ${\bf x} \mapsto F( {\bf x})$ w.r.t. to the variable ${\bf x}$ at time step $k$, i.e.
$$ {\bf g}_{k} = \nabla_{\bf x} F( {\bf x}_{k} ).$$
The SGD update is given by $ {\bf x}_{k+1} = {\bf x}_{k} - \eta \, {\bf g}_{k}.$ In order to improve the convergence, Netwon's method can be used. Then the update is given by 

$${\bf x}_{k+1} = {\bf x}_{k} - \eta \, [ \nabla^{2} F({\bf x}_{k}))]^{-1} \nabla_{\bf x} F( {\bf x}_{k} ).$$

Computing the inverse of Hessian in the Newton's method is expensive and are replaced by preconditioner $B$ in methods such as BFGS, i.e.

$${\bf x}_{k} = {\bf x}_{k} - \eta \, B^{-1} \nabla_{\bf x} F( {\bf x}_{k} ).$$

Let use define

$$ B_{k} = \sqrt{ {\rm diag} \left( \sum_{s=1}^{k} [\nabla F( {\bf x}_{s} )] [\nabla F( {\bf x}_{s} )]^{\rm T} \right)}.$$

In AdaGrad, the learning rate $\eta$ is updated
$$ {\bf w} = {\bf w} - \eta \, {\rm diag}(G)^{-1/2} \circ {\bf g}$$
or componentwise, 
$${\bf w}_{k+1, i} = {\bf w}_{k, i} - \dfrac{\eta}{\sqrt{G_{k, ii} + \epsilon}}  g_{k, i}$$
where
$$ G_{k} = \sum_{j=1}^{k} {\bf g}_{j} {\bf g}_{j}^{\rm T}.$$

## Reference
 
[1] J Duchi, E Hazan, and Y Singer, "Adaptive subgradient methods for online learning and stochastic optimization," Journal of Machine Learning Research 12, pp. 2121-2159, 2011.

```python
###################### Adaptive gradient ###############

def adaptive_gradient(X, y, eta=0.01, max_iterations=10):
    eps = 1e-08
    m, n = X.shape
    L, W, G = np.zeros( max_iterations ), np.zeros((n,max_iterations-1)), np.zeros((n,n))
    w_m = 5*np.random.rand(n,1)
    W[:,:0] = w_m
    
    for k in range(max_iterations):
        cost = 0.0
        w_0 = w_m
       
        # Reshuffle the data (IMPORTANT!)
        r = np.random.permutation(m)
        X, y = X[r], y[r]
        
        for i in range(m):
            xi = X[i,:].reshape(n,1)     # row vector
            yi = y[i]                    # scalar
            g = (sum(xi[:]*w_m[:]) - yi)* xi
            G += np.outer( g,g )
            DG = G.reshape(n,n)
            z = np.diag(np.sqrt(np.diag(DG))) + eps
            z = inv(z)
            w_m = w_m - eta* z.dot(g)
            cost += (1/2)*(np.square( sum(xi[:]*w_m[:]) - yi))
        
        L[k]  = cost.ravel()/m
        W[:,:k] = w_m

        
    return W, L

W_AdaGrad, L_AdaGrad = adaptive_gradient(Xaug, y, eta, max_iterations)
Result.append( ['ADAGRAD', W_AdaGrad, L_AdaGrad] )
```

# (h) Adaptive Moment Estimation (Adam)

Adam combines momentum and RMSProp.

## Algorithm

-  Step size $\eta$
-  $\beta_{1}, \beta_{2} \in [0,1)$ -  Exponential decay rates for the moment estimates
-  $L( {\bf w};Z)$: Stochastic objective function with parameters $\bf w$
-  Initialize ${\boldsymbol \theta}_{0} \in \mathbb{R}^{n+1}$: Initial parameter vector 
-  ${\bf v}_{0} \leftarrow {\bf 0} \in \mathbb{R}^{n+1}$ (Initialize 1st moment vector)
-  ${\bf s}_{0} \leftarrow {\bf 0} \in \mathbb{R}^{n+1}$ (Initialize 2nd moment vector)
-  $k  \leftarrow 0$ (Initialize timestep)
-  while ${\boldsymbol \theta}_{k}$ not converged do
    -  $k \leftarrow k+ 1$
    -  ${\bf g}_{k} \leftarrow \nabla_{\boldsymbol \theta} L( {\boldsymbol \theta}_{k−1}; Z)$
    -  ${\bf v}_{k} \leftarrow \beta_{1} {\bf v}_{k−1} + (1 - \beta_{1}) \, {\bf g}_{k}$ ("momentum" step)
    -  ${\bf s}_{k} \leftarrow \beta_{2} {\bf s}_{k−1} + (1 - \beta_{2}) \, ( {\bf g}_{k}\circ {\bf g}_{k})$ ("RMSProp" step)
    -  $\widehat{\bf s}_{k} \leftarrow \displaystyle{ \frac{ {\bf m}_{k}}{(1 − [\beta_{1}]^{k})}}$ (Compute bias-corrected first moment estimate)
    -  $\widehat{ \bf v}_{k} \leftarrow \displaystyle{ \frac{ {\bf v}_{k} }{ (1 − [\beta_{2}]^{k} )}}$ (Compute bias-corrected second raw moment estimate)
    -  ${\bf w}_{k} \leftarrow {\bf w}_{k-1} - \displaystyle{ \frac{ \eta \widehat{\bf s}_{k}}{\sqrt{ \widehat{\bf v}_{k}} + \epsilon} }$ (Update parameters) 
-  end while

## Reference
[1] DP Kingma and JL Ba, "ADAM: A method for stochastic optimization," ICLR 2015.

```python
###################### Adaptive Moment Estimation (ADAM) ###############

def adam(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    eps = 1e-08
    beta1, beta2 = 0.9, 0.999
    v0, s0 = np.zeros((n,1)), np.zeros((n,1))
    L, W, G = np.zeros(max_iterations), np.zeros((n,max_iterations-1)), np.zeros((n,max_iterations-1))
    w0 = 5*np.random.rand(n,1)
    W[:,:0] = w0
    
    for k in range(max_iterations):
        g1 = np.dot( X.transpose(),(X.dot(w0) - y.reshape(m,1))) / m

        v1 = beta1 * v0 + (1-beta1) * g1
        s1 = beta2 * s0 + (1 - beta2) * (g1 ** 2)
        
        hat_v1 = v1/(1 - beta1 ** (k+1))
        hat_s1 = s1/(1 - beta2 ** (k+1))
       
        w1 = w0 - eta * hat_v1/(np.sqrt(hat_s1) + eps)

        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w1) - y.reshape(m,1)) )
        W[:,:k] = w1
        G[:,:k] = g1
        
        w0, s0, v0 = w1, s1, v1

    return W, L, G

W_Adam, L_Adam, G_Adam = adam(Xaug, y, eta, max_iterations)
Result.append( ['ADAM', W_Adam, L_Adam] )
```

# (i) AMSGrad

AMSGrad that uses the maximum of past squared gradients ${\bf v}_{t}$ rather than the exponential average to update the parameters.

## Algoirthm

-  Stepsize $\eta$
-  $\beta_{1}, \beta_{2} \in [0,1)$
-  Exponential decay rates for the moment estimates
-  $L( {\bf w};X, {\bf y})$: Stochastic objective function with parameters $\bf w$
-  Initialize ${\bf w}_{0} \in \mathbb{R}^{n+1}$: Initial parameter vector 
-  ${\bf m}_{0} \leftarrow {\bf 0} \in \mathbb{R}^{n+1}$ (Initialize 1st moment vector)
-  ${\bf v}_{0} \leftarrow {\bf 0} \in \mathbb{R}^{n+1}$ (Initialize 2nd moment vector)
-  $k  \leftarrow 0$ (Initialize timestep)
-  while ${\bf w}_{k}$ not converged do
    -  $k \leftarrow k+ 1$
    -  ${\bf g}_{k} \leftarrow \nabla_{\bf w} L( {\bf w}_{k−1}; X, {\bf y})$ (Get gradients w.r.t. stochastic objective at timestep $k$)
    -  ${\bf m}_{k} \leftarrow \beta_{1} {\bf m}_{k−1} + (1 - \beta_{1}) {\bf g}_{k}$ (Update biased first moment estimate)
    -  ${\bf v}_{k} \leftarrow \beta_{2} {\bf v}_{k−1} + (1 - \beta_{2}) {\bf g}_{k}\circ {\bf g}_{k}$ (Update biased second raw moment estimate)
    -  $\hat{\bf v}_{k} = \text{max}(\hat{\bf v}_{k-1}, {\bf v}_{k}) \\
$ (Compute bias-corrected second raw moment estimate)
    -  ${\bf w}_{k} \leftarrow {\bf w}_{k-1} - \displaystyle{ \frac{ \eta }{\sqrt{ \hat{\bf v}_{k}} + \epsilon} } {\bf m}_{k}$ (Update parameters) 
-  end while
-  return ${\bf w}_{k}$ (Resulting parameters)

## Reference

SJ Reddi, S Kale, and S Kumar, ``On the Convergence of Adam and Beyond,`` Proceedings of ICLR 2018. 

```python
###################### Adaptive Moment Estimation (ADAM) ###############

def AMSGrad(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    eps = 1e-08
    beta1 = 0.9
    beta2 = 0.999
    mt = np.zeros( (n,1) )
    vt = np.zeros( (n,1) )
    L = np.zeros(max_iterations)
    W = np.zeros((n,max_iterations-1))
    w = 5*np.random.rand(n,1)
    hat_vt = np.ones( (n,1) )
    W[:,:0] = w


    for k in range(max_iterations):
        gt = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
        mt = beta1*mt + (1-beta1)*gt
        vt = beta2*vt + (1 - beta2)*np.multiply(gt,gt)
        hat_vt = np.maximum(vt,hat_vt)
        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
        w = w - eta * mt/(np.sqrt(hat_vt) + eps)
        W[:,:k] = w

    return W, L

W_AMSGrad, L_AMSGrad = AMSGrad(Xaug, y, eta, max_iterations)
Result.append( ['AMSGrad', W_AMSGrad, L_AMSGrad] )
```

# (j) AdaDelta

Adadelta is an extension of Adagrad which reduces its aggressive, monotonically decreasing learning rate $\eta$. Instead accumulating all past squared gradients ${\bf g}_{i} \circ {\bf g}_{i}$, $ 0 \leq i \leq k,$ Adadelta restricts the window of accumulated past gradients to some fixed size $0 < l \leq i \leq k .$

\begin{align*}
{\bf u}_{k + 1} & = \gamma \, {\bf u}_{k} + (1 - \gamma) \left( \nabla L({\boldsymbol \theta}_k) \circ \nabla L({\boldsymbol \theta}_k) \right) \\
{\bf v}_{k + 1} & = -\frac{\sqrt{ {\bf x}_{k} + \epsilon} }{\sqrt{ {\bf u}_{k+1} + \epsilon}} \nabla L(\theta_k) \\
{\bf w}_{k + 1} & = \gamma {\bf x}_{k} + (1 - \gamma) {\bf v}_{k + 1}^2 \\
{\boldsymbol \theta}_{k + 1} & = {\boldsymbol \theta}_{k} + {\bf v}_{k + 1}
\end{align*}


## Reference

[1] MD Zeiler, "ADADELTA: An Adaptive Learning Rate Method," 2012 (http://arxiv.org/abs/1212.5701).

```python
###################### AdaDelta ###################

def AdaDelta(X, y, eta=0.01, max_iterations=10):
    m, n  = X.shape
    eps   = 1e-06
    gamma = 0.95
    L, W  = np.zeros(max_iterations), np.zeros((n,max_iterations-1))
    w     = 5*np.random.rand(n,1)
    W[:,:0] = w
   
    Eg = np.zeros((n,1))
    Edw = np.zeros((n,1))
    
    for k in range(max_iterations):
        g = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
        Eg = gamma * Eg + (1 - gamma)*(g*g)
      
        dw = (np.sqrt(Edw + eps)/np.sqrt(Eg + eps))*g
        
        Edw = gamma*Edw + (1-gamma)*(dw*dw)
        w = w - eta*dw
                
        L[k] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)))
        W[:,:k] = w
        
    return W, L
    
W_AdaDelta, L_AdaDelta = AdaDelta(Xaug, y, eta, max_iterations)
Result.append( ['ADADELTA', W_AdaDelta, L_AdaDelta] )
plt.plot(range(max_iterations),L_AdaDelta,'-' )
```

# (k) AdaMax


## Reference

[1] DP Kingma and JL Ba, "Adam: A method for stochastic optimization," International Conference on Learning Representations, pp. 1–13, 2015.

```python
###################### Adaptive Moment Estimation (ADAM) ###############

def AdaMax(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    eps = 1e-08
    beta1 = 0.9
    beta2 = 0.999
    mt = np.zeros( (n,1) )
    vt = np.zeros( (n,1) )
    L = np.zeros(max_iterations)
    W = np.zeros((n,max_iterations-1))
    w = 5*np.random.rand(n,1)
    hat_vt = np.ones( (n,1) )
    W[:,:0] = w
    gamma = 0.999   
    g = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
    L[0] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
    Eg = np.square(g)

    for i in range(max_iterations):
                
        g = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
    
        mt = (beta1 * mt + (1 - beta1) * g)
        vt = (beta2 * vt + (1 - beta2) * (g ** 2))
    
        ut = np.max(np.append(beta2 * vt, g, axis=0), axis=0)
    
        mhat = mt / (1 - beta1 ** (i+1) ) 
        vhat = vt / (1 - beta2 ** (i+1) )
    
        W[:,:i] = w
        L[i] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
        
        w = w-eta * mhat/(np.sqrt(ut) + eps)
    
    return W, L

W_AdaMax, L_AdaMax = AdaMax(Xaug, y, eta, max_iterations)
Result.append( ['ADAMAX', W_AdaMax, L_AdaMax] )
```

```python
def Nadam(X, y, eta=0.01, max_iterations=10):
    m, n = X.shape
    eps = 1e-08
    beta1 = 0.9
    beta2 = 0.999
    mt = np.zeros( (n,1) )
    vt = np.zeros( (n,1) )
    L = np.zeros(max_iterations)
    W = np.zeros((n,max_iterations-1))
    w = 5*np.random.rand(n,1)
    hat_vt = np.ones( (n,1) )
    W[:,:0] = w
    gamma = 0.999
    g = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
    L[0] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
    Eg = np.square(g)

    for i in range(max_iterations):

        g = np.dot( X.transpose(),(X.dot(w) - y.reshape(m,1))) / m
    
        mt = (beta1 * mt + (1 - beta1) * g)
        vt = (beta2 * vt + (1 - beta2) * (g ** 2))
    
        ut = np.max(np.append(beta2 * vt, g, axis=0), axis=0)
    
        mhat = mt / (1 - beta1 ** (i+1)) 
        vhat = vt / (1 - beta2 ** (i+1))
  
        w = w - eta * (beta1 * mhat + (1 - beta1)/(1 - beta1 ** (i+1)) * g)/(np.sqrt(ut) + eps)
        L[i] = (1/(2*m))*np.sum( np.square(X.dot(w) - y.reshape(m,1)) )
        W[:,:i] = w
    
    return W, L

W_Nadam, L_Nadam = Nadam(Xaug, y, eta, max_iterations)
Result.append(['NADAM', W_Nadam, L_Nadam])
```

# Direct method of optimization

Direct search methods are those which rely on function evaulations $F({\bf x})$ on a sequence $\{ {\bf x}_{n} \}$ and comparing values in order to calculate a minimizer of $F.$ This approach is usually applied to instances such as 
non differentiable $F( \cdot)$, complicated derivatives, etc. The most popular one are

*  Golden section method (applicable to single variable function)
*  Fibonacci method (applicable to single variable function)
*  Hooke and Jeeves’ method
*  Sendley, Hext and Himsworth’s method
*  Nelder and Mead’s method

## Hooke and Jeeve's Method

Consider the minimization problem

$$ \min_{ {\bf x}\in \mathbb{R}^{n}} F( {\bf x})$$

The method proceeds by a sequence of exploratory and pattern moves from a base point. If an exploratory move leads to a decrease of the value of f(x) it is called a success; otherwise it is called a failure.

#### Exploratory moves

1.  Choose a base point ${\bf b}_{1} \in \mathbb{R}^{n}$ and step length $h_{1}.$
2.  for $i = 1$ to $n$ do
    - If $F( {\bf b}_{i} + h_{i} {\bf e}_{i}) < F( {\bf b}_{i} )$, then replace ${\bf b}_{i}$ by ${\bf b}_{i} + h_{i} {\bf e}_{i}.$ 
    - Else, if $F( {\bf b}_{i} - h_{i} {\bf e}_{i})< F( {\bf b}_{i} )$, replace ${\bf b}_{i}$ by ${\bf b}_{i} - h_{i} {\bf e}_{i}.$
    - Else retain ${\bf b}_{i}.$
    
3. If ${\bf b}_{n} = {\bf b}_{1},$ halve each of the step lengths $h_{j}$ and return to Step 1. Terminate when the lengths have been reduced to the prescribed precision.

#### pattern moves $( {\bf b}_{n} \neq {\bf b}_{1})$

1.  Move from ${\bf b}_{n}$ to ${\bf p}_{1}= 2 {\bf b}_{n} − {\bf b}_{1}$ and continue with a new sequence of
exploratory moves about ${\bf p}_{1}.$

2.  If the lowest function value obtained during the pattern and exploratory moves of Step 1 is less than
$F(b2),$ then a new base point ${\bf b}_{3}$ has been reached. Return to Step 1 with all suffices increased by 1. Otherwise, abandon the pattern move from b2 and continue with a new sequence of exploratory moves.
